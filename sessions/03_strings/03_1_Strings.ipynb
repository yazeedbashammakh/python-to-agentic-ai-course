# Strings in Python

So far, we've worked with numbers.

But many real-world programs need to work with text:

- Names
- Email addresses
- Cities
- Messages
- User input

In Python, text is represented using **strings**.

## Creating Strings

Strings can be created using either single quotes (`' '`) or double quotes (`" "`).

````python
'hello'
````

````python
"This is also a string"
````

Both approaches create strings.

Choose whichever makes your text easier to write and read.

````python
"I'm learning Python"
````

````python
'Now we are learning "Python"'
````

## Displaying Strings

In Jupyter notebooks, simply writing a string will display it.

However, the preferred way to display output is using `print()`.

````python
print("Hello World")
print("Hello")
print("World")
````

### New Lines

Use `\n` to move to a new line.

````python
print("Python\nAI\nEngineering")
````

## Measuring String Length

Python provides the `len()` function to count characters in a string.

Spaces count as characters too.

````python
len("Hello World")
````

## Strings Are Sequences

A string is a sequence of characters.

Each character has a position called an **index**.

````python
s = "Hello World"
s
````

Python starts counting from **0**.

````python
s[0]
````

````python
s[1]
````

````python
s[6]
````

## Slicing Strings

Slicing allows us to extract part of a string.

Syntax:

```python
string[start:end]
```

Python includes the start index but excludes the end index.

````python
s[1:]
````

````python
s[:5]
````

````python
s[3:8]
````

````python
s[:]
````

## Negative Indexing

Negative indices count from the end of the string.

````python
s[-1]
````

````python
s[-2]
````

````python
s[-5]
````

## String Immutability

Strings cannot be modified after they are created.

This property is called **immutability**.

````python
s[0] = "X"
````

The code above raises an error because individual characters cannot be changed.

## Concatenation

Strings can be joined together using the `+` operator.

````python
"Hello" + " World"
````

````python
first_name = "Rohan"
last_name = "Sharma"

first_name + " " + last_name
````

## Repetition

The `*` operator can repeat strings.

````python
"Python " * 3
````

## Common String Methods

````python
s.upper()
````

````python
s.lower()
````

````python
"python programming".title()
````

````python
"   hello   ".strip()
````

````python
"Hello World".replace("World", "Python")
````

## String Formatting

String formatting allows us to insert values into strings.

````python
name = "Rohan"

print("Hello, my name is " + name)
````

Modern Python uses **f-strings**, which are cleaner and easier to read.

````python
name = "Rohan"

print(f"Hello, my name is {name}")
````
