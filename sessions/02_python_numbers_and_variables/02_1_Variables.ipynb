# Variables in Python

So far, we've worked directly with numbers.

```python
10 + 5
```

This works well for simple calculations, but real programs need a way to store information and reuse it later.

That's where **variables** come in.

Variables are one of the most important concepts in programming and will be used throughout this course. 

## Creating Your First Variable

A variable allows us to store a value and give it a name.

The syntax is:

```python
variable_name = value
```

The `=` symbol is called the **assignment operator** because it assigns a value to a variable.

````python
a = 5
````

Once a value is assigned, we can use the variable name to access that value.

````python
a
````

Variables can be used anywhere a value can be used.

a + 5

## Updating a Variable

Variables are not permanent.

We can assign a new value to the same variable name.

````python
a = 10
a
````

The old value is replaced with the new value.

````python
a + a
````

## Reassigning Variables

Sometimes we want to update a variable based on its current value.

Python allows us to reference the variable while assigning a new value.

````python
maths = 89

# recheck paper --> 1 questions m aap galat the and aapke 3 marks kate jate h

maths = maths - 3

a = a + 10
a
````

Let's break this down:

1. Take the current value of `a`
2. Add `10`
3. Store the result back into `a`

## Shortcut Operators

Updating variables is very common.

Python provides shorthand operators for this purpose.

````python
a += 10
a
````

The statement above is equivalent to:

```python
a = a + 10
```

````python
a *= 2
a
````

Similarly:

```python
a -= 5
a /= 2
```

can be used for subtraction and division.

# Real-World Example

Variables become useful when we need meaningful names.

Compare:

```python
100 * 0.1
```

with

```python
income * tax_rate
```

The second version is much easier to understand.

my_income = 100
tax_rate = 0.1 

my_taxes = my_income * tax_rate

print(my_taxes)

## Checking Data Types

Every value in Python has a type.

We can inspect the type using the built-in `type()` function.

````python
type(a)
````

Currently, `a` contains a whole number, so Python treats it as an integer (`int`).

a = 2.5
type(a)

Now `a` contains a decimal value, so Python treats it as a float (`float`).

Using descriptive variable names makes programs easier to read and maintain.
