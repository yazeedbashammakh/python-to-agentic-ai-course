# List Comprehension

Throughout the previous notebooks, we've used **for loops** to process data and create new lists.

Most of the time, the process looked something like this:

1. Create an empty list.
2. Loop through another collection.
3. Add new values using `append()`.

Let's look at a familiar example.

````python
students = ["Rohan", "Mohan", "Chandu", "Ajay", "Vijay"]

students_copy = []

for student in students:
    students_copy.append(student)

students_copy
````

This works perfectly.

But notice that most of the code is simply setting up the loop.

Python provides a shorter and cleaner way to create new lists from existing collections.

This is called **List Comprehension**.

### First List Comprehension

````python
students_copy = [student for student in students]

students_copy
````

The syntax may look different at first, but it's doing exactly the same work.

```
new_list = [expression for item in iterable]
```

Where:

- **expression** → The value to add to the new list.
- **item** → The current element being processed.
- **iterable** → The collection being looped over.

### Recognizing the Pattern

````python
# Example 1

numbers = [1, 2, 3, 4]

# Traditional
squares = []

for number in numbers:
    squares.append(number ** 2)

print(squares)
````

↓

````python
squares = [number ** 2 for number in numbers]

print(squares)
````

````python
# Example 2

students = ["Rohan", "Mohan", "Chandu"]

# Traditional
student_caps = []

for student in students:
    student_caps.append(student.upper())

print(student_caps)
````

↓

````python
student_caps = [student.upper() for student in students]

print(student_caps)
````

### Filtering

Sometimes we don't want every item.

We only want the items that satisfy a condition.

In that case, place the `if` condition at the end.

````python
marks = [80, 39, 52, 95, 28]

passed_marks = [mark for mark in marks if mark >= 40]

passed_marks
````

### Conditional Modification

Sometimes we want to keep every item but change its value based on a condition.

In that case, use an `if-else` expression before the `for` keyword.

````python
marks = [80, 39, 52, 95, 28]

results = [
    "Pass" if mark >= 40 else "Fail"
    for mark in marks
]

results
````

### Real-world Examples

````python
# Example 1 Strings
students = ["Rohan", "Mohan", "Chandu"]

student_caps = [student.upper() for student in students]

student_caps
````

````python
# Example 2 Dictionaries
student_marks = {
    "Rohan": 83,
    "Mohan": 91,
    "Ajay": 48
}

student_names = [student for student in student_marks]

student_names
````

Then

````python
marks = [mark for mark in student_marks.values()]

marks
````
