```python
marks = [85, 75, 65, 25, 38]

status = []
for mark in marks:
  if mark > 40:
    status.append('Passed')
  else:
    status.append('Failed')
print(status)
```

```python
Rohan_marks = [83, 75, 89]
Mohan_marks = [89, 91, 93]
Chandu_marks = [65, 78, 74]
Ajay_marks = [46, 52, 48]
Vijay_marks = [70, 80, 98]


```

```python
def grade_student(marks):
  total = sum(marks)
  pcct = total/len(marks)

  if pcct > 90:
    grade = 'A+'
  elif pcct > 80:
    grade = 'A'
  elif pcct > 70:
    grade = 'B'
  elif pcct > 60:
    grade = 'C'
  else:
    grade = 'F'
  print(f"Total Marks: {total}")
  print(f"pcct: {pcct}")
  print(f"grade: {grade}")

  return [total, pcct, grade]
```

```python
grade_student(Rohan_marks)
grade_student(Mohan_marks)
grade_student(Chandu_marks)
grade_student(Ajay_marks)
grade_student(Vijay_marks)
```
