## The Problem

Suppose we have created a function to display a student's information.

Every time we call the function, we need to provide the values in the correct order.

````python
def student_info(name, age, city):
    print("Name :", name)
    print("Age  :", age)
    print("City :", city)
````

````python
student_info("Rohan", 20, "Pune")
````

This works perfectly.

But what happens if we accidentally change the order of the values?

Let's try it.

````python
student_info(20, "Pune", "Rohan")
````

The function still runs successfully because Python assigns the values based on their position.

Since the values are passed in the wrong order, the output is also incorrect.

Wouldn't it be better if we could tell Python exactly which value belongs to which parameter?

This is where **Keyword Arguments** become useful.

## Keyword Arguments

Instead of relying on the order of the values, we can explicitly specify the parameter name while calling the function.

This makes the function call easier to read and prevents mistakes caused by passing values in the wrong order.

````python
student_info(
    age=20,
    city="Pune",
    name="Rohan"
)
````

Notice that the order of the arguments no longer matters because each value is assigned using its parameter name.

When values are passed according to their position, they are called **Positional Arguments**.

When values are passed using parameter names, they are called **Keyword Arguments**.

Both approaches are valid, and you'll see both used in real-world Python programs.

## Default Parameters

Python allows us to assign a default value to a parameter.

If no value is provided while calling the function, Python automatically uses the default value.

````python
def student_info(name, age, city="Pune"):
    print("Name :", name)
    print("Age  :", age)
    print("City :", city)
````

````python
student_info("Rohan", 20)
````

````python
student_info("Ajay", 21, "Mumbai")
````

Default parameters help reduce repetitive code while still allowing us to provide a different value whenever needed.

They are especially useful when a function usually receives the same value for a particular parameter.

## The Next Problem

So far, every function we've written returns only one value.

For example, suppose we calculate a student's result.

Is returning only the total marks enough?

Probably not.

We may also need:

- Total Marks
- Percentage
- Grade

Do we need to create three separate functions for this?

````python
def student_result(maths, science, english):

    total = maths + science + english

    return total
````

````python
total = student_result(83, 75, 89)

print(total)
````

This works, but we still need to calculate the percentage and grade somewhere else.

Instead of returning one value, Python allows a function to return multiple related values together.

````python
def student_result(maths, science, english):

    total = maths + science + english
    percentage = total / 300 * 100

    if percentage >= 90:
        grade = "A"

    elif percentage >= 75:
        grade = "B"

    elif percentage >= 60:
        grade = "C"

    else:
        grade = "D"

    return total, percentage, grade
````

````python
result = student_result(83, 75, 89)

print(result)
````

The function returned three values together.

Notice that the output is enclosed inside parentheses `()`.

> 🚀 **Did You Know?**
>
> Python automatically groups multiple returned values into a collection called a **Tuple**.
>
> A Tuple is similar to a List, but it is generally used for storing related values that belong together.
>
> You don't need to learn Tuples separately for this course, but it's useful to know that Python creates one automatically whenever a function returns multiple values.

Although the function returns multiple values together, we can also store each value in its own variable.

````python
total, percentage, grade = student_result(83, 75, 89)

print("Total      :", total)
print("Percentage :", round(percentage, 2))
print("Grade      :", grade)
````

This technique is called **unpacking** because Python automatically separates each returned value into its corresponding variable.

You'll use this pattern quite often while working with APIs, AI libraries, and many built-in Python functions.

## The Next Problem

Sometimes, returning a few values isn't enough.

Suppose we have marks for multiple students, and we only want the names of the students who passed.

Returning one name at a time wouldn't be very useful.

Instead, we can collect all the passed students into a list and return the entire list.

````python
students = {
    "Rohan": 83,
    "Mohan": 91,
    "Ajay": 48,
    "Chandu": 74,
    "Vijay": 35
}
````

Let's create a function that returns the names of all students who scored **50 or more**.

````python
def passed_students(student_marks):

    passed = []

    for student, marks in student_marks.items():

        if marks >= 50:
            passed.append(student)

    return passed
````

````python
result = passed_students(students)

print(result)
````

Notice that our function didn't return a single value.

Instead, it returned an entire list.

Since Lists are objects in Python, they can be returned from functions just like numbers, strings, or dictionaries.

In fact, many Python functions and libraries return lists because they allow us to work with multiple pieces of related data at once.

````python
passed = passed_students(students)

print("Passed Students:")

for student in passed:
    print(student)
````

Returning collections such as Lists or Dictionaries makes functions much more reusable.

Instead of printing data directly inside the function, we return the data and decide later how we want to use it.

This gives us much greater flexibility while building larger applications.

> 💡 **Programming Insight**
>
> Notice that we simply call `passed_students()` and immediately get the result.
>
> We don't need to know how the function loops through the dictionary or checks each student's marks.
>
> The function hides those implementation details and only gives us the final result.
>
> Hiding unnecessary implementation details while exposing only what the user needs is known as **Abstraction**.
>
> You don't need to study Abstraction in detail for this course, but it's a useful programming concept you'll often encounter in documentation and software development.

## The Next Problem

So far, every function we've created expects a fixed number of parameters.

For example, our `student_result()` function expects marks for exactly three subjects.

But what if another class has:

- 4 subjects?
- 5 subjects?
- 8 subjects?

Do we need to rewrite the function every time the number of subjects changes?

````python
def calculate_total(maths, science, english):

    return maths + science + english
````

````python
print(calculate_total(83, 75, 89))
````

This works perfectly for three subjects.

However, if we try to pass marks for four or five subjects, Python raises an error because our function only accepts three parameters.

Instead of limiting our function to a fixed number of values, Python allows us to accept as many values as needed.

## *args

The `*args` syntax allows a function to receive any number of positional arguments.

This makes our function much more flexible because it no longer depends on a fixed number of parameters.

````python
def calculate_total(*marks):

    total = 0

    for mark in marks:
        total += mark

    return total
````

````python
print(calculate_total(83, 75, 89))
````

````python
print(calculate_total(83, 75, 89, 92, 88))
````

Notice that we didn't change the function.

Instead, the same function automatically handled both three and five marks.

This makes our code much more reusable because the function can adapt to different situations without being rewritten.

> 💡 **Programming Insight**
>
> Notice how the same function works with different numbers of inputs without us having to create a new function each time.
>
> Writing code that can adapt to different situations instead of rewriting similar code is one of the basic ideas behind a programming concept called **Polymorphism**.
>
> You don't need to study Polymorphism in detail for this course, but it's a useful term to recognize because you'll often come across it in Python documentation, books, and discussions with other developers.

## The Next Problem

The `*args` syntax works well when we only care about the values.

For example, we can calculate the total marks without knowing which subject each mark belongs to.

However, sometimes the labels are just as important as the values.

For example, instead of passing:

```python
83, 75, 89
```

it would be much more meaningful to pass:

```python
Maths=83
Science=75
English=89
```

Python allows us to do this using `**kwargs`.

## **kwargs

The `**kwargs` syntax allows a function to receive any number of **named arguments**.

Each value is passed together with its name, making the function calls much easier to understand.

````python
def student_marks(**marks):

    for subject, mark in marks.items():
        print(subject, ":", mark)
````

````python
student_marks(
    Maths=83,
    Science=75,
    English=89
)
````

Unlike `*args`, where we only receive values, `**kwargs` allows us to receive both the names and their corresponding values.

This makes functions much more descriptive and easier to understand, especially when working with large amounts of related data.

````python
def student_report(name, **marks):

    print("Student :", name)

    total = 0

    for subject, mark in marks.items():
        print(subject, ":", mark)
        total += mark

    print("Total :", total)
````

````python
student_report(
    "Rohan",
    Maths=83,
    Science=75,
    English=89
)
````

Notice how the function became much easier to use.

Instead of remembering the order of the marks, we simply provide each subject name along with its value.

This makes our functions more readable and reduces the chances of passing incorrect data.

> 💡 **Programming Insight**
>
> Notice that our `student_report()` function performs multiple related tasks:
>
> - Displays the student's name
> - Displays each subject and its marks
> - Calculates the total
> - Displays the final total
>
> All of this related logic is kept together inside a single function.
>
> Organizing related code into one place is one of the basic ideas behind a programming concept called **Encapsulation**.
>
> You don't need to study Encapsulation separately for this course, but it's a useful term to recognize because you'll often encounter it in programming books, documentation, and discussions.

## The Next Problem

So far, every function we've created has used the `def` keyword.

This works well for functions that contain multiple lines of code.

But sometimes a function performs only a single operation.

Writing an entire function for such a small task can feel unnecessary.

Python provides a shorter way to create simple functions using **Lambda Functions**.

## Lambda Functions

A Lambda Function is a small anonymous function that is usually written in a single line.

It is commonly used for simple calculations or operations where creating a full function would be unnecessary.

Syntax:

```python
lambda parameters: expression
```

````python
# Normal Function

def square(number):
    return number ** 2

print(square(5))
````

The same function can be written using a Lambda Function.

````python
square = lambda number: number ** 2

print(square(5))
````

Notice that the lambda function performs exactly the same task.

The main difference is that it is written in a single line.

For larger or more complex functions, using `def` is usually a better choice because it is easier to read and maintain.

Lambda functions are most useful for short and simple operations.

````python
students = ["Rohan", "Mohan", "Ajay", "Chandu"]

students.sort(key=lambda name: len(name))

print(students)
````

In this example, the lambda function tells Python to sort the student names based on their length instead of alphabetical order.

You'll often see lambda functions used with built-in functions such as:

- `sort()`
- `sorted()`
- `map()`
- `filter()`

For now, it's enough to recognize the syntax and understand why lambda functions exist.

As you continue learning Python, you'll naturally encounter more situations where they are useful.
