# Functions

So far in this course, we've learned how to:

- Store data using variables
- Organize data using lists and dictionaries
- Make decisions using conditional statements
- Repeat tasks using loops

Using these concepts, we can already build many useful programs.

But as our programs become larger, a new problem starts to appear.

Let's discover what that problem is.

## The Problem

Suppose we want to calculate the total marks of every student in a class.

We already know how to solve this using a `for` loop.

````python
Rohan_marks = [83, 75, 89]
Mohan_marks = [89, 91, 93]
Chandu_marks = [65, 78, 74]
Ajay_marks = [46, 52, 48]
Vijay_marks = [70, 80, 98]
````

````python
Rohan_total = 0

for mark in Rohan_marks:
    Rohan_total += mark

print(Rohan_total)
````

This works perfectly.

Now let's calculate the total marks for another student.

````python
Mohan_total = 0

for mark in Mohan_marks:
    Mohan_total += mark

print(Mohan_total)
````

Now imagine doing the same thing for every student in the class.

````python
Chandu_total = 0

for mark in Chandu_marks:
    Chandu_total += mark

print(Chandu_total)


Ajay_total = 0

for mark in Ajay_marks:
    Ajay_total += mark

print(Ajay_total)


Vijay_total = 0

for mark in Vijay_marks:
    Vijay_total += mark

print(Vijay_total)
````

Can you spot the problem?

The logic for calculating the total marks is exactly the same every time.

Only the student's marks are changing.

Imagine writing this code for:

- 50 students
- 500 students
- 5000 students

The code quickly becomes repetitive and difficult to maintain.

Now imagine your school changes the grading system and you need to update the calculation.

You would have to find every copy of the same code and modify it.

There should be a better way.

Instead of writing the same logic again and again, wouldn't it be better if we could write it **once** and reuse it whenever we need it?

This is exactly why Python provides **Functions**.

## Creating Our First Function

Instead of writing the same logic for every student, we can write it **once** and give it a name.

Whenever we need the same logic again, we simply use its name instead of rewriting the code.

This reusable block of code is called a **Function**.

The basic syntax of a function is:

```python
def function_name():
    # Code to execute
```

Here:

- `def` is a keyword used to define a function.
- `function_name` is the name we give to the function.
- The indented code contains the instructions that belong to the function.

````python
def hello():
    print("Hello World")
````

Run the code above.

Did anything happen?

Nothing was printed.

That's because we have only **defined** the function.

A function definition simply tells Python **what to do**.

It does **not** execute the code automatically.

To execute the function, we need to **call** it.

````python
hello()
````

Now the message is printed.

This is called **calling** (or **invoking**) the function.

You can call the same function as many times as you want without rewriting the code.

````python
hello()
hello()
hello()
````

## Function Definition vs Function Call

Every function has two parts.

### Function Definition

This is where we write the instructions.

```python
def hello():
    print("Hello World")
```

### Function Call

This is where we ask Python to execute those instructions.

```python
hello()
```

Think of it this way:

- **Definition** → Creating the recipe.
- **Call** → Following the recipe to prepare the dish.

You only define a function once, but you can call it as many times as needed.

## Making Functions More Flexible

Our `hello()` function always prints the same message.

```python
Hello World
```

But what if we want to greet different students?

Should we create a separate function for every student?

````python
def hello_rohan():
    print("Hello Rohan")

def hello_mohan():
    print("Hello Mohan")

def hello_ajay():
    print("Hello Ajay")
````

This works...

But we've created the same function three times.

Only the student's name changes.

Just like before, we're repeating our code again.

Instead, we can make the function accept the student's name while calling it.

````python
def greet(name):
    print("Hello", name)
````

Here, `name` is a **parameter**.

A parameter is simply a variable that receives a value when the function is called.

````python
greet("Rohan")
greet("Mohan")
greet("Ajay")
greet("Vijay")
````

The values we pass while calling the function are called **arguments**.

For example,

```python
greet("Rohan")
```

- `name` → Parameter
- `"Rohan"` → Argument

Every time we call the function, the parameter receives a new value, allowing the same function to work with different data.

Functions are useful because the logic stays the same while the data changes.

Instead of writing multiple functions like:

- `hello_rohan()`
- `hello_mohan()`
- `hello_ajay()`

we only need one reusable function:

```python
greet(name)
```

This makes our code shorter, cleaner, and much easier to maintain.

Greeting students is useful, but functions can do much more than printing messages.

Suppose we want to calculate the sum of two numbers.

````python
print(34 + 76)
````

This works.

But what if we want to calculate the sum of many different numbers?

Instead of writing a new calculation every time, we can reuse a function.

````python
def add_numbers(a, b):
    print(a + b)
````

## Returning Values

Our `add_numbers()` function prints the answer.

```python
add_numbers(25, 30)
```

Output:

```
55
```

This is useful if we only want to display the result.

But what if we want to **store** the result in a variable and use it later?

````python
total = add_numbers(25, 30)

print(total)
````

You might expect the output to be:

```
55
55
```

But instead you'll get something like:

```
55
None
```

Why did this happen?

The function printed the answer, but it never **returned** it.

Printing a value and returning a value are two different things.

- `print()` displays information on the screen.
- `return` sends a value back to wherever the function was called.

````python
def add_numbers(a, b):
    return a + b
````

Now the function sends the result back instead of printing it.

````python
total = add_numbers(25, 30)

print(total)
````

Since the function returns a value, we can use it anywhere an ordinary value can be used.

````python
print(add_numbers(10, 15))
````

````python
result = add_numbers(40, 20)

print(result * 2)
````

````python
final_total = (
    add_numbers(20, 30) +
    add_numbers(10, 40)
)

print(final_total)
````

## `print()` vs `return`

Although both seem similar at first, they serve different purposes.

| `print()` | `return` |
|-----------|----------|
| Displays the value | Sends the value back |
| Used for output | Used for further processing |
| Cannot be reused later | Can be stored in variables and reused |

As your programs become larger, you'll use `return` much more often than `print()`.
