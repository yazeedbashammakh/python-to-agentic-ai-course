# Booleans in Python

So far, we've worked with:

- Numbers
- Strings
- Variables

Now we'll learn about another important data type called **Boolean**.

A Boolean can have only one of two values:

- `True`
- `False`

Booleans are used whenever we need to represent a yes/no, on/off, or true/false condition.

## Creating Booleans

Python provides two built-in Boolean values:

- `True`
- `False`

Notice that the first letter is capitalized.

```python
a = True

a
```

```python
b = False

b
```

## Why Do We Need Booleans?

Many real-world situations have only two possible states.

Examples:

- Logged in or not logged in
- Payment completed or not completed
- Email verified or not verified
- Light switched on or off

Booleans allow us to represent these situations in code.

```python
is_logged_in = True

is_logged_in
```

```python
payment_completed = False

payment_completed
```

## Creating Booleans Using Comparisons

Comparison operations often produce Boolean values.

The result will always be either:

- `True`
- `False`

```python
1>2
```

Since 1 is not greater than 2, Python returns `False`.

```python
10 > 5
```

```python
3 == 3
```

```python
7 != 2
```

Don't worry about memorizing comparison operators right now.

We'll cover them in detail later when we learn conditional statements.

```python
is_greater = 1 > 2

is_greater

type(is_greater)
```

This is a common pattern in programming.

We often store conditions in variables and use them later in our programs.

## Introducing None

Sometimes we need a variable, but we don't yet know its value.

In such cases, Python provides a special value called `None`.

```python
user_name = None

user_name
```

`None` means:

> "No value has been assigned yet."

```python
type(user_name)
```

Notice that `None` is not a Boolean.

It has its own special type called `NoneType`.

```python
print(user_name)
```

## Real-World Example

Imagine a website where a user has not yet entered their email address.

```python
email = None

email
```

Later, when the user provides an email address, we can update the variable.

```python
email = "student@example.com"

email
```
