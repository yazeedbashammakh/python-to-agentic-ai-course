# Lists in Python

So far, we've learned how to store information using variables.

For example:

```python
student_1 = "Rohan"
student_2 = "Mohan"
student_3 = "Chandu"
student_4 = "Ajay"
student_5 = "Vijay"
```

This works, but becomes difficult to manage as the amount of data grows.

Imagine storing information for:

- 50 students
- 500 students
- 5000 students

Instead of creating hundreds of variables, Python provides **Lists**.

A List allows us to store multiple values inside a single variable.

## Creating a List

Lists are created using square brackets `[]`.

Each element is separated by a comma.

````python
students = ["Rohan", "Mohan", "Chandu", "Ajay", "Vijay"]

students
````

Each value inside the list is called an **element**.

## Length of a List

We can use the `len()` function to count how many elements exist inside a list.

````python
len(students)
````

## Accessing Elements (Indexing)

Lists use indexing just like strings.

Python starts counting from index `0`.

````python
students[0]
````

````python
students[2]
````

````python
students[-1]
````

## Slicing Lists

Slicing allows us to extract multiple elements from a list.

Syntax:

```python
list[start:end]
```

The start index is included.

The end index is excluded.

````python
students[1:]
````

````python
students[:3]
````

````python
students[1:4]
````

## Combining Lists

Lists can be joined together using the `+` operator.

````python
students + ["Sohan"]
````

Notice that the original list does not change.

````python
students
````

To permanently update the list, we must reassign it.

````python
students = students + ["Sohan"]

students
````

## Repeating Lists

The `*` operator can duplicate list elements.

````python
students * 2
````

## Adding Elements

Use `append()` to add an element to the end of a list.

````python
students.append("Raju")

students
````

Use `insert()` to add an element at a specific position.

````python
students.insert(2, "Rohan")

students
````

## Removing Elements

Use `pop()` to remove elements.

````python
students.pop()
````

````python
students
````

````python
students.pop(0)
````

````python
students
````

## Removing Specific Elements

Use `remove()` when you know the value you want to delete.

````python
students.remove("Rohan")

students
````

## Clearing a List

Use `clear()` to remove all elements from a list.

````python
students.clear()

students
````

````python
students = ["Rohan", "Mohan", "Chandu", "Ajay", "Vijay"]

students
````

## Searching Inside Lists

````python
"Rohan" in students
````

````python
"Raj" in students
````

## Finding Position of an Element

````python
students.index("Vijay")
````

## Counting Occurrences

````python
students.count("Vijay")
````

## Sorting and Reversing Lists

````python
students.reverse()

students
````

````python
students.sort()

students
````
