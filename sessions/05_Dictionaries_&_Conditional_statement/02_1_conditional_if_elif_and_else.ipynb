# Conditional Statements

So far, every Python program we've written has followed the same pattern.

Python starts from the first line and executes every statement one by one until it reaches the end of the program.

For example,

````python
print("Wake Up")
print("Brush Teeth")
print("Go to College")
print("Study Python")
````

Every time you run this program, Python produces exactly the same output.

It doesn't stop.

It doesn't make any decisions.

It simply follows the instructions from top to bottom.

This is how every program we've written has behaved so far.

But think about real life.

Our daily decisions are rarely this straightforward.

## The Problem

Imagine a school wants to display the exam result of a student.

The rule is simple.

- If the student scores **40 or more**, they pass.
- Otherwise, they fail.

Can we write this using only what we've learned so far?

````python
total_marks = 85

print("Passed")
print("Failed")
````

Something doesn't look right.

The student cannot both **pass** and **fail**.

Python printed both messages because we told it to execute both lines.

It doesn't know that only one of them should be displayed.

We need a way to tell Python:

> Execute a piece of code **only when a certain condition is satisfied.**

This is exactly why Python provides the **if statement**.

## The `if` Statement

The `if` statement allows Python to execute a block of code only when a condition is true.

If the condition is false, Python simply skips that block.

Syntax

```python
if condition:
    # code to execute
```

Let's solve our exam problem properly.

````python
total_marks = 85

if total_marks >= 40:
    print("Passed")
````

Since the student scored **85**, the condition is satisfied.

Python enters the `if` block and prints:

```
Passed
```

Now try changing the value of `total_marks` to **35** and run the program again.

What happened?

````python
total_marks = 35

if total_marks >= 40:
    print("Passed")
````

Nothing was printed.

Python checked the condition.

Since it was false, it skipped the code inside the `if` block and moved on.

This is useful because we can now execute code **only when we want to**.

But another problem appears.

## Another Problem

If the student doesn't pass, shouldn't the program display **Failed**?

Right now, nothing happens.

```python
total_marks = 35

if total_marks >= 40:
    print("Passed")
```

The student failed the exam, but the program doesn't tell us that.

Wouldn't it be nice if Python could automatically execute another block whenever the condition is false?

That's exactly what the `else` statement does.

## The `else` Statement

The `else` statement is used together with `if`.

If the `if` condition is true, Python executes the `if` block.

Otherwise, it executes the `else` block.

Syntax

```python
if condition:
    # code
else:
    # code
```

Let's improve our program.

````python
total_marks = 85

if total_marks >= 40:
    print("Passed")
else:
    print("Failed")
````

Now change the marks to different values and run the program again.

For example:

- 25
- 40
- 65
- 90

Notice something important.

Only **one** message is printed.

If the `if` block executes, Python skips the `else` block.

If the `if` condition is false, Python skips the `if` block and executes the `else` block.

This makes the program behave much more like real life.

## One More Problem

Suppose your school doesn't just display **Pass** or **Fail**.

Instead, it awards grades.

The rules are:

- Less than 40 → Failed
- 40–59 → Grade D
- 60–69 → Grade C
- 70–79 → Grade B
- 80–89 → Grade A
- 90 and above → Grade A+

Can we solve this using only `if` and `else`?

Technically, yes.

But the code quickly becomes difficult to read because there are many different possibilities.

Python provides a cleaner solution.

The **`elif` statement**.

It allows us to check multiple conditions one after another until one of them becomes true.

## The `elif` Statement

The `elif` statement allows Python to check **multiple conditions**.

Instead of asking just one question, Python can ask a series of questions until it finds the correct answer.

The syntax looks like this:

```python
if condition:
    # code
elif condition:
    # code
else:
    # code
```

Python starts from the top and checks each condition one by one.

As soon as one condition becomes **True**, Python executes that block and ignores the remaining conditions.

Let's solve our grading problem.

````python
total_marks = 85

if total_marks < 40:
    grade = "Failed"
elif total_marks < 60:
    grade = "Grade D"
elif total_marks < 70:
    grade = "Grade C"
elif total_marks < 80:
    grade = "Grade B"
elif total_marks < 90:
    grade = "Grade A"
else:
    grade = "Grade A+"

print(grade)
````

Since the student scored **85**, Python checks the conditions one by one.

```
85 < 40   ❌ False
85 < 60   ❌ False
85 < 70   ❌ False
85 < 80   ❌ False
85 < 90   ✅ True
```

Once Python finds a condition that is **True**, it executes that block and stops checking the remaining conditions.

That's why the output is:

```
Grade A
```

Try changing the value of `total_marks` to different values and observe the output.

For example,

- 25
- 40
- 65
- 72
- 89
- 95

Notice that only **one grade** is printed every time.

## Why Does Python Stop Checking?

Imagine a student scores **95**.

Should the student receive:

- Grade D
- Grade C
- Grade B
- Grade A
- Grade A+

Of course not.

The student should receive only **one** grade.

That's why Python stops checking as soon as it finds the first matching condition.

Let's see another example.

````python
temperature = 32

if temperature < 10:
    print("Very Cold")
elif temperature < 20:
    print("Cold")
elif temperature < 30:
    print("Pleasant")
else:
    print("Hot")
````

Changing a single variable changes the entire output.

Try these values:

- 5
- 15
- 25
- 40

Observe which condition becomes true.

## One Important Rule

The **order of conditions matters**.

Python always checks the conditions from **top to bottom**.

If the correct condition appears later, Python may never reach it.

Consider this example.

````python
total_marks = 95

if total_marks >= 40:
    print("Passed")
elif total_marks >= 90:
    print("Grade A+")
````

You might expect the output to be:

```
Grade A+
```

But the actual output is:

```
Passed
```

Why?

Because Python starts checking from the top.

The very first condition,

```python
total_marks >= 40
```

is already **True**.

So Python executes that block and skips everything below it.

The `elif` statement is never checked.

A better way to write this program is to place the more specific conditions first.

````python
total_marks = 95

if total_marks >= 90:
    print("Grade A+")
elif total_marks >= 40:
    print("Passed")
else:
    print("Failed")
````

Now the student correctly receives:

```
Grade A+
```

Whenever you're writing multiple conditions, always think about the order in which Python will evaluate them.

## Indentation

One thing you've probably noticed throughout this notebook is the spacing before each statement.

This spacing is called **indentation**.

Unlike many programming languages, Python uses indentation to identify which statements belong to each block.

Every statement inside an `if`, `elif`, or `else` block must have the same indentation.

````python
age = 20

if age >= 18:
    print("Eligible to Vote")
    print("You can cast your vote.")
else:
    print("Not Eligible")

print("Program Finished")
````

Notice that both `print()` statements inside the `if` block are indented.

This means they both belong to the `if` block.

The last `print()` statement has no indentation, so it is outside the conditional statement and always executes.

Now look at the following code.

````python
# This code will produce an IndentationError

age = 20

if age >= 18:
print("Eligible")
````

Python doesn't know which statements belong to the `if` block because the indentation is missing.

Always remember to indent every line inside an `if`, `elif`, or `else` block.
