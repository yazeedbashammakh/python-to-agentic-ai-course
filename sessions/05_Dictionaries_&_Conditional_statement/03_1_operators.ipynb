# Comparison Operators

In the previous section, we learned how Python makes decisions using `if`, `elif`, and `else`.

For example,

```python
marks = 85

if marks >= 40:
    print("Passed")
else:
    print("Failed")
```

This program works perfectly.

But have you noticed something?

How does Python know whether **85 is greater than or equal to 40**?

How does it decide whether the condition is **True** or **False**?

The answer lies in **Comparison Operators**.

Comparison operators compare two values and always produce a Boolean value:

- `True`
- `False`

These Boolean values are what allow conditional statements to make decisions.

## Comparing Values

Let's see a few examples.

Try to predict the output before running each cell.

````python
5 == 5
````

````python
5 == 8
````

````python
10 != 3
````

````python
10 != 10
````

````python
15 > 8
````

````python
15 < 8
````

````python
20 >= 20
````

````python
18 <= 10
````

Notice something?

Every comparison produces either:

```python
True
```

or

```python
False
```

That's because comparison operators answer simple questions such as:

- Are these values equal?
- Are they different?
- Is one value greater than the other?
- Is one value smaller than the other?

These Boolean values are then used by `if` statements to decide which block of code should execute.

## Assignment vs Comparison

This is one of the most common mistakes beginners make.

Look carefully at these two operators.

```python
=
```

and

```python
==
```

Although they look similar, they perform completely different tasks.

````python
marks = 85
````

Here,

`=` assigns a value to a variable.

````python
marks == 85
````

Here,

`==` compares two values.

It asks the question:

> "Is the value of `marks` equal to 85?"

The result will be either `True` or `False`.

Always remember:

- `=` → Assign a value
- `==` → Compare two values

## Comparison Operators with Strings

Comparison operators don't work only with numbers.

They can also compare strings.

````python
"Rohan" == "Rohan"
````

````python
"Rohan" == "Mohan"
````

````python
"Python" != "Java"
````

Just like numbers, Python compares the values and returns either `True` or `False`.

## Membership Operators

Suppose a classroom contains the following students.

````python
students = [
    "Rohan",
    "Mohan",
    "Ajay",
    "Vijay",
    "Chandu"
]
````

Now imagine we want to know whether **Rohan** belongs to this class.

Instead of checking every name one by one, Python provides **Membership Operators**.

````python
"Rohan" in students
````

````python
"Mayank" in students
````

````python
"Mayank" not in students
````

The `in` operator checks whether a value exists inside a collection such as a list.

The `not in` operator checks whether a value does **not** exist.

Just like comparison operators, these also return either `True` or `False`.

## Combining Multiple Conditions

So far, every decision we've made has been based on a single condition.

For example,

```python
marks >= 40
```

But real-world problems often require checking **more than one condition**.

Imagine a scholarship is awarded only if a student scores more than **80** in both Maths and Science.

How can Python check both conditions together?

This is where **Logical Operators** become useful.

````python
maths = 88
science = 90

(maths > 80) and (science > 80)
````

The `and` operator returns `True` only when **both conditions** are true.

````python
maths = 88
science = 70

(maths > 80) or (science > 80)
````

The `or` operator returns `True` if **at least one** condition is true.

````python
students = ["Rohan", "Mohan", "Ajay"]

"Mayank" not in students
````

The `not` operator reverses the result.

If the original condition is `True`, it becomes `False`.

If it is `False`, it becomes `True`.

## Using Logical Operators with Conditional Statements

Now let's combine everything we've learned so far.

````python
age = 65

if age < 5 or age > 60:
    print("Free Ticket")
else:
    print("Regular Ticket")
````

Python first evaluates the condition.

```python
age < 5 or age > 60
```

The result is either `True` or `False`.

That result is then passed to the `if` statement, which decides which block of code should execute.
