# Dictionaries in Python

So far, we've learned how to store information using:

- Variables → Store a single value
- Lists → Store multiple values

But what if we want to store **related information with meaningful labels**?

Let's understand this with a real-world example.

## The Problem

Suppose we want to store marks for a student.

The student has marks in three subjects:

- Maths
- Science
- English

One way to store this information is by creating separate variables.

````python
maths = 60
science = 70
english = 80

total_marks = maths + science + english

print(total_marks)
````

This works perfectly for **one student**.

But imagine storing marks for:

- 50 students
- 500 students
- 5000 students

Creating separate variables for every student's subject quickly becomes difficult to manage.

Can Lists solve this problem?

## Using a List

We can store the marks inside a list.


````python
marks = [60, 70, 80]

marks
````

This is better.

But there is one problem.

Can you immediately tell which subject has **70 marks**?

```python
marks[1]
```

You might remember today that index `1` is Science, but after a few weeks—or if another developer reads your code—it becomes difficult to understand.

Instead of remembering index positions, wouldn't it be better if we could access the marks using the subject name?

This is exactly why Python provides **Dictionaries**.

## Creating a Dictionary

A Dictionary stores information as **key-value pairs**.

Think of a key as a label and the value as the information stored under that label.

Syntax:

```python
{
    key: value
}
```

````python
rohan_marks = {
    "maths": 60,
    "science": 70,
    "english": 80
}

rohan_marks
````

Instead of remembering index positions, we can now access marks using the subject name.

````python
rohan_marks["science"]
````

````python
rohan_marks["maths"]
````

## Updating Values

Just like variables, dictionary values can also be updated.

````python
rohan_marks["science"] = rohan_marks["science"] + 10

rohan_marks
````

Python also supports the shortcut assignment operator.

````python
rohan_marks["english"] += 5

rohan_marks
````

## Adding New Information

Dictionaries are dynamic.

We can add new keys whenever needed.

````python
rohan_marks["computer"] = 95

rohan_marks
````

## Removing Information

Use the `del` keyword to remove a key-value pair.

````python
del rohan_marks["computer"]

rohan_marks
````

## Dictionary Methods

Python provides several useful methods for working with dictionaries.

### Getting All Keys

````python
rohan_marks.keys()
````

### Getting All Values

````python
rohan_marks.values()
````

### Getting Both Keys and Values

````python
rohan_marks.items()
````

## Checking if a Key Exists

Sometimes we only want to know whether a key exists inside the dictionary.

````python
"maths" in rohan_marks
````

````python
"history" in rohan_marks
````

## Safely Accessing Values

If a key doesn't exist, directly accessing it will raise an error.

Instead, we can use the `get()` method.

````python
rohan_marks.get("history")
````

## List vs Dictionary

Both Lists and Dictionaries store collections of data, but they are used differently.

| List | Dictionary |
|------|------------|
| Uses indexes | Uses keys |
| Ordered collection | Key-value mapping |
| `students[0]` | `marks["maths"]` |
| Best for sequences | Best for labeled data |

A good rule of thumb:

- Use a **List** when order matters.
- Use a **Dictionary** when labels matter.

## Nested Dictionaries

Now imagine we need to store marks for an entire class.

Each student has their own dictionary.

This creates a **Nested Dictionary**.

````python
student_marks = {
    "rohan": {
        "maths": 60,
        "science": 70,
        "english": 80
    },
    "mohan": {
        "maths": 70,
        "science": 80,
        "english": 90
    },
    "chandu": {
        "maths": 80,
        "science": 50,
        "english": 90
    }
}

student_marks
````

To access nested data, simply continue indexing using keys.

````python
student_marks["rohan"]["maths"]
````

````python
student_marks["mohan"]["english"]
````

## Dictionaries with Lists

Sometimes the values inside a dictionary are also lists.

This is another common way of organizing data.

````python
student_marks = {
    "rohan": [60, 70, 80],
    "mohan": [70, 80, 90],
    "chandu": [80, 50, 90]
}

student_marks
````

Here:

- Index 0 → Maths
- Index 1 → Science
- Index 2 → English

````python
student_marks["rohan"][0]
````
