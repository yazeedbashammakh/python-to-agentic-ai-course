```python
marks = [79, 85, 93, 25, 67]
total_marks = 0

for mark in marks:
  total_marks += mark

print(f"total_marks:{total_marks}")
```

```python
marks = [79, 85, 93, 25, 67]

for mark in marks:
  if mark < 30:
    print(f"{mark} - Failed")
  else:
    print(f"{mark} - Passed")
```

```python
marks = [79, 85, 93, 25, 67]

for mark in marks:
  if mark < 30:
    print(f'{mark}: Grade: F')
  elif mark < 60:
    print(f'{mark}: Grade: C')
  elif mark < 70:
    print(f'{mark}: Grade: B')
  elif mark < 80:
    print(f'{mark}: Grade: A')
  else:
    print(f'{mark}: Grade: A+')
```

```python
marks = [79, 85, 93, 25, 67]

updated_marks = []
for mark in marks:
  updated_marks.append(mark + 5)

print(updated_marks)

```

```python
marks = [79, 85, 93, 25, 67]

passed_marks = []
for mark in marks:
  if mark > 30:
    passed_marks.append(mark)

passed_marks
```

```python
marks = [79, 85, 93, 25, 67]

for i in range(len(marks)):
  print(f"index {i}: {marks[i]}")
```

```python
student ={
    "name": "Rohan",
    "age": 12,
    "city": "Pune"
}

for key in student:
  print(f"{key}: {student[key]}")
```

```python
for key in student.keys():
  print(f"{key}: {student[key]}")
```

```python
for value in student.values():
  print(f"{value}")
```

```python
for key, value in student.items():
  print(f"{key}: {value}")
```

```python
marks = [85, 75, 65, 45, 13]

i = 0
while i < len(marks):
  print(marks[i])
  i += 1
```

```python
marks = [85, 75, 65, 45, 13]

i = 0
while i < len(marks):
  print(marks[i])
  i += 1

else:
  print(f"loop breaked at {i}")
```
