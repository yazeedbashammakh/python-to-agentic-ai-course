# While Loops

In the previous session, we learned how to repeat tasks using **for loops**.

A `for` loop is one of the most useful tools in Python.

But does it solve every repetition problem?

Let's find out.

## The Problem

Suppose we want to print the numbers from **1 to 5**.

Since we've already learned `for` loops, we can easily solve it like this.

````python
for number in range(1, 6):
    print(number)
````

This works perfectly.

The `for` loop already knows where to start and where to stop.

But now imagine a different situation.

Suppose we want to keep repeating a task until a condition changes.

For example:

- Keep asking for a password until it is correct.
- Keep processing orders until there are no orders left.
- Keep counting visitors until 100 people arrive.

In these situations, we don't always know beforehand how many times the loop should run.

Can a `for` loop solve every one of these problems?

Instead of repeating a fixed number of times, wouldn't it be better if we could repeat code **while a condition remains True?**

This is exactly why Python provides **While Loops**.

## Creating Our First While Loop

A `while` loop repeats a block of code **as long as its condition remains True**.

Syntax:

```python
while condition:
    # code
```

````python
count = 1

while count <= 5:
    print(count)
    count += 1
````

Notice something important.

Unlike a `for` loop, a `while` loop does **not** automatically move to the next value.

We must update the variable ourselves.

Once the condition becomes `False`, the loop stops.

Let's use another example.

This time, let's count backwards.

````python
count = 5

while count > 0:
    print(count)
    count -= 1

print("Blast Off!")
````

## Using While Loops with Lists

Earlier, we used a `for` loop to process every item in a list.

Can a `while` loop solve the same problem?

Absolutely.

The only difference is that, instead of automatically moving through the list, we keep track of the position ourselves using an **index**.

````python
students = ["Rohan", "Mohan", "Chandu", "Ajay", "Vijay"]

index = 0

while index < len(students):
    print(students[index])
    index += 1
````

Notice how the loop works.

- We start from index `0`.
- After printing one student, we increase the index.
- The loop stops once the index reaches the length of the list.

The output is exactly the same as our `for` loop.

Only the way we move through the list is different.

Printing values is useful, but loops are often used to process data.

Let's solve a more practical problem.

## Processing Student Marks

We'll continue using the same student records from the previous notebook.

This time, let's calculate the total marks for every student using a `while` loop.

````python
student_marks = {
    "Rohan": {
        "Maths": 83,
        "English": 75,
        "Science": 89
    },
    "Mohan": {
        "Maths": 89,
        "English": 91,
        "Science": 93
    },
    "Chandu": {
        "Maths": 65,
        "English": 78,
        "Science": 74
    },
    "Ajay": {
        "Maths": 46,
        "English": 52,
        "Science": 48
    },
    "Vijay": {
        "Maths": 70,
        "English": 80,
        "Science": 98
    }
}

students = list(student_marks.keys())

index = 0

while index < len(students):

    student = students[index]
    marks = student_marks[student]

    total = (
        marks["Maths"] +
        marks["English"] +
        marks["Science"]
    )

    print(student, ":", total)

    index += 1
````

Just like the `for` loop, we processed every student one by one.

The only extra work is maintaining the index ourselves.

Let's build on this one more time.

Instead of displaying only the total marks, let's also calculate each student's percentage.

````python
students = list(student_marks.keys())

index = 0

while index < len(students):

    student = students[index]
    marks = student_marks[student]

    total = (
        marks["Maths"] +
        marks["English"] +
        marks["Science"]
    )

    percentage = total / 300 * 100

    print(student, "-", round(percentage, 2), "%")

    index += 1
````

So far, we've only displayed information.

Can we also make decisions while the loop is running?

Of course.

Just like a `for` loop, a `while` loop can also work together with `if`, `elif`, and `else`.

## Making Decisions Inside a While Loop

So far, we've processed every student's marks and calculated their percentage.

But in a real application, simply displaying percentages isn't enough.

Schools usually assign grades based on the percentage.

Can we make these decisions while the `while` loop is running?

Absolutely.

Just like a `for` loop, a `while` loop can also work together with `if`, `elif`, and `else`.

````python
students = list(student_marks.keys())

index = 0

while index < len(students):

    student = students[index]
    marks = student_marks[student]

    total = (
        marks["Maths"] +
        marks["English"] +
        marks["Science"]
    )

    percentage = total / 300 * 100

    if percentage >= 90:
        grade = "A"

    elif percentage >= 75:
        grade = "B"

    elif percentage >= 60:
        grade = "C"

    else:
        grade = "D"

    print(student, "-", round(percentage, 2), "%", "-", grade)

    index += 1
````

Now our loop is doing much more than printing data.

For every student it:

- Calculates the total marks.
- Calculates the percentage.
- Decides the grade.
- Displays the final result.

This is how loops are commonly used in real-world programs—they process data and make decisions at the same time.

Let's make the report a little more informative.

Instead of displaying only the grade, let's also determine whether each student has **Passed** or **Failed**.

````python
students = list(student_marks.keys())

index = 0

while index < len(students):

    student = students[index]
    marks = student_marks[student]

    total = (
        marks["Maths"] +
        marks["English"] +
        marks["Science"]
    )

    percentage = total / 300 * 100

    if percentage >= 50:
        result = "Pass"
    else:
        result = "Fail"

    print(student, "-", round(percentage, 2), "%", "-", result)

    index += 1
````

Our loop is working exactly as expected.

It starts from the beginning and continues until the condition becomes `False`.

But what if we want to change that behavior?

Can we stop the loop early?

Can we skip certain iterations?

Let's explore a few situations where we need more control over a `while` loop.

## Controlling a While Loop

In the previous session, we learned how `break`, `continue`, and `pass` work inside a `for` loop.

The same keywords can also be used inside a `while` loop.

Let's see them in action.

### Using `break`

Sometimes we don't want to process every item.

Instead, we want to stop the loop as soon as a certain condition is met.

````python
students = ["Rohan", "Mohan", "Chandu", "Ajay", "Vijay"]

index = 0

while index < len(students):

    student = students[index]

    if student == "Chandu":
        print(f"{student} Found!")
        break

    print(student)

    index += 1
````

### Using `continue`

Instead of stopping the loop, we can simply skip the current iteration and continue processing the remaining students.

````python
students = ["Rohan", "Mohan", "Chandu", "Ajay", "Vijay"]

index = 0

while index < len(students):

    student = students[index]

    index += 1

    if student == "Ajay":
        continue

    print(student)
````

### Using `pass`

Just like in a `for` loop, `pass` can be used as a placeholder while developing your program.

````python
students = ["Rohan", "Mohan", "Chandu", "Ajay", "Vijay"]

index = 0

while index < len(students):

    student = students[index]

    if student == "Chandu":
        pass

    print(student)

    index += 1
````

## The `while...else` Statement

Unlike the previous examples, this is a feature specific to `while` loops.

The `else` block runs only when the loop finishes normally.

````python
students = ["Rohan", "Mohan", "Chandu", "Ajay", "Vijay"]

index = 0

while index < len(students):
    print(students[index])
    index += 1

else:
    print("All students processed.")
````

````python
students = ["Rohan", "Mohan", "Chandu", "Ajay", "Vijay"]

index = 0

while index < len(students):

    student = students[index]

    if student == "Chandu":
        print("Student Found!")
        break

    print(student)

    index += 1

else:
    print("All students processed.")
````

## One Common Mistake

Every example we've written so far updates the variable used in the condition.

But what happens if we forget to do that?

````python
count = 1

while count <= 5:
    print(count)
````

Can you spot the problem before running the code?

Since `count` never changes, the condition always remains `True`.

As a result, the loop never ends.

This is called an **Infinite Loop**.
